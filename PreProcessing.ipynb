```python
import json
import numpy as np

def get_data(file_path):
    with open(file_path, 'r') as file:
        data = file.read()

    # Step 2: Parse the JSON data
    parsed_data = json.loads(data)

    # Step 3: Access the data
    word_seq = parsed_data['word_seq']
    
    return word_seq
```

```python
file_path = "Right.txt"
data = get_data(file_path)
```

```python
data_x = np.array(data["x"])
data_y = np.array(data["y"])
time_diff = np.array(data["time"])
```

```python
def process_time(time):
    ret=[]
    total = 0
    for val in time:
        total += val
        ret.append(total)
    return np.array(ret)
```

```python
def preprocess_data(x, y, time_diff):

    # makes the first x and y points 0,0
    x = x - x[0]
    y = y - y[0]
    time = process_time(time_diff)

    return x, y, time
```

```python
x, y, time = preprocess_data(data_x, data_y, time_diff)
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1,2)
axs[0].plot(x,y*-1)
axs[1].plot(data_x, data_y*-1)
plt.show
```
